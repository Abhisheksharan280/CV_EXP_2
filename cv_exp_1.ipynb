```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
import matplotlib
import numpy as np
import pandas as pd
from PIL import Image
import imutils
```

```python
image1 = cv2.imread('/content/img1.jpg')
cv2_imshow(image1)
```

```python
r1 = cv2.resize(image1, (255,255))
cv2_imshow(r1)
```

```python
input1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
(input1)
```

```python
cv2_imshow(input1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image2 = cv2.imread('/content/img2.png')
cv2_imshow(image2)
```

```python
r2 = cv2.resize(image2, (255,255))
cv2_imshow(r2)
```

```python
#ADD
add1 = cv2.add(r1,r2)
cv2_imshow(add1)
```

```python
#sub
sub1 = cv2.subtract(r1,r2)
cv2_imshow(sub1)
```

```python
image = cv2.imread("./content/img1.jpg")

Rotated_image = imutils.rotate(image, angle=45)
Rotated1_image = imutils.rotate(image, angle=90)

cv2.imshow("Rotated 45 degrees", Rotated_image)
cv2.imshow("Rotated 90 degrees", Rotated1_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
