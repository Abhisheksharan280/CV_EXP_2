```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.io import imshow
```

```python
image1 = cv2.imread('/content/IMG1.png')
plt.imshow(image1)
```

```python
ret,mask=cv2.threshold(image1,50,255,cv2.THRESH_BINARY)
plt.imshow(mask)
kernel= np.ones((5,5),np.uint8)
```

```python
dilation=cv2.dilate(mask,kernel,iterations=2)
plt.imshow(dilation)
```

```python
erosion=cv2.erode(mask,kernel,iterations=4)
plt.imshow(erosion)
```

```python
dilation=cv2.dilate(erosion,kernel,iterations=2)
plt.imshow(dilation)
```

```python
image2=cv2.imread('/content/IMG2 (2).png')
plt.imshow(image2)
```

```python
ret,mask=cv2.threshold(image2,50,255,cv2.THRESH_BINARY)
plt.imshow(mask)
kernel= np.ones((5,5),np.uint8)
```

```python
opening=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)
plt.imshow(opening)
```

```python
closing=cv2.morphologyEx(image2,cv2.MORPH_CLOSE,kernel)
plt.imshow(closing)
```
