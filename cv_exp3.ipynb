```python
import cv2
import matplotlib.pyplot as plt
import skimage
from skimage.feature import graycomatrix, graycoprops
from skimage import data
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
s = cv2.imread('/content/img1.jpg',cv2.IMREAD_GRAYSCALE)
c = cv2.imread('/content/img2.jpg',cv2.IMREAD_GRAYSCALE)
r = cv2.imread('/content/img3.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
cv2_imshow(s)
```

```python
cv2_imshow(c)
```

```python
cv2_imshow(r)
```

```python
#Smooth Picture Details -
s_co = skimage.feature.graycomatrix(s,[5],[0],levels=256, symmetric=True, normed=True)
s_contrast= skimage.feature.graycoprops(s_co,'contrast')
s_corel = skimage.feature.graycoprops(s_co,'correlation')
s_energy = skimage.feature.graycoprops(s_co,'energy')
s_homo = skimage.feature.graycoprops(s_co,'homogeneity')

```

```python
print("Contrast = ", s_contrast )
print("Correlation = ", s_corel )
print("Energy = ", s_energy )
print("Homogeneity = ", s_homo )
```

```python
#Coarse Picture Details -
c_co = skimage.feature.graycomatrix(c,[5],[0],levels=256, symmetric=True, normed=True)
c_contrast= skimage.feature.graycoprops(c_co,'contrast')
c_corel = skimage.feature.graycoprops(c_co,'correlation')
c_energy = skimage.feature.graycoprops(c_co,'energy')
c_homo = skimage.feature.graycoprops(c_co,'homogeneity')

print("Contrast = ", c_contrast )
print("Correlation = ", c_corel )
print("Energy = ", c_energy )
print("Homogeneity = ", c_homo )
```

```python
#Coarse Picture Details -
r_co = skimage.feature.graycomatrix(c,[5],[0],levels=256, symmetric=True, normed=True)
r_contrast= skimage.feature.graycoprops(r_co,'contrast')
r_corel = skimage.feature.graycoprops(r_co,'correlation')
r_energy = skimage.feature.graycoprops(r_co,'energy')
r_homo = skimage.feature.graycoprops(r_co,'homogeneity')

print("Contrast = ", r_contrast )
print("Correlation = ", r_corel )
print("Energy = ", r_energy )
print("Homogeneity = ", r_homo )
```
